```python

from typing import List



def removeDuplicates(nums: List[int]) -> int:
    num = [0] * 201
    removal = []
    count=0
    for i in range(len(nums)):
        if(num[nums[i]+100] == 0):
            count +=1;
            num[nums[i]+100] += 1
        else:
            removal.append(i)

    for i in range(len(removal)-1,-1,-1):
        del nums[removal[i]]
    return count;

nums = [100] * (3*(10 ** 4))

expectedNums = [100]; # The expected answer with correct length

k = removeDuplicates(nums); # Calls your implementation
# print(nums)
assert k == len(expectedNums);

for i in range(k):
    assert nums[i] == expectedNums[i];

# print(removeDuplicates(nums))
```
