$$
a^2 + b^2 = c^2
$$

```python
import itertools as it
for a,b,c in it.combinations(range(0,1000),3):
    if (a ** 2 + b ** 2) == c **2:
        print(f'a:{a},b:{b},c:{c}')
        exit();
```

Sum of powers conjecture
In 1769, Euler conjectured that for any integer n >2 one needs at least n summands to represent the nth power of a positive integer as a sum of the nth powers of positive integers. In particular, equations

$a^4 +b^4 + c+4 = d^4 \text{ and } a^5 +b^5 + c+5 + d^5= e^5$

```python
# No solution exist
import itertools as it5
for a,b,c,d,e in it.combinations(range(100,200),5):
    if (a ** 5 + b ** 5 + c **5 + d ** 5) == e **5:
        print(f'a:{a},b:{b},c:{c}')
        exit();
```
